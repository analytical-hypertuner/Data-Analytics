categorical plots

```
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```
tips = sns.load_dataset('tips')
tips.head()
```

```
#bar plot n count plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```
sns.countplot(x='sex',data=tips)
```

```
sns.countplot(x ='sex', hue = "smoker", data = tips)
 
# Show the plot
plt.show()
```

```
#boxplot
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

```
#violinplot
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow',hue='sex')
```

```
#strip plot
sns.stripplot(x="day", y="total_bill", data=tips)
```

```
#swarmplot
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1")
```

```
# Combining Categorical Plots
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

```
#factorplot
sns.factorplot(x='sex',y='total_bill',data=tips,kind='box')
```

```
#catplot
sns.catplot(x="day", y="total_bill", kind="strip", data=tips)
```

```
#relplot
sns.relplot(data=tips, x="total_bill", y="tip", hue="day",kind="scatter") # scatter n lineplot
```

```
#Distribution plots
#displot
sns.displot(tips['total_bill'],kde=True,bins=8)
```

```
sns.displot(tips['total_bill'])
```

```
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="stack")
```

```
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```
#pairplot
sns.pairplot(tips,kind="hist",hue='sex')
```

```
iris = sns.load_dataset("iris")
g = sns.pairplot(iris)
plt.show()
```

```
#rugplot
sns.rugplot(tips['total_bill'])
```

```
# marix,grid n regression plot
```

```
tips.corr()
```

```
#matrix plot
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```

```
#Grid plot
#facetgrid
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.scatter, "total_bill","tip").add_legend()
```

```
#jointplot
sns.jointplot(x="total_bill",y="tip",data=tips,kind='hex')
plt.show()
```

```
#regression plots
sns.regplot(x="total_bill", y="tip", data=tips)
```

```

import numpy as np 
import matplotlib.pyplot as plt 
  
fig = plt.figure()
ax = plt.axes(projection ='3d') 

# defining axes 
z = np.linspace(0, 1, 100) 
x = z * np.sin(25 * z) 
y = z * np.cos(25 * z) 



ax.plot3D(x, y, z, 'green')   #lineplot
ax.set_title('3D line plots ') 
plt.show() 
```

```
 
import numpy as np 
import matplotlib.pyplot as plt 
  
fig = plt.figure()
ax = plt.axes(projection ='3d') 

# defining axes 
z = np.linspace(0, 1, 100) 
x = z * np.sin(25 * z) 
y = z * np.cos(25 * z) 

c=x+y
ax.scatter(x,y,z,c=c)

ax.set_title('3D scatter plots ') 
plt.show() 
```
